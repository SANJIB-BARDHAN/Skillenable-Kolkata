Import the required libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the dataset into a Pandas DataFrame:


```python
df = pd.read_excel(r'E:\SKILLENABLE\INTERVIEW PREP\Olympic Athletes.xlsx')
print(df)

```

Task I: Which countries have won the most medals?


```python
most_medals = df.groupby('Country')['Total Medals'].count().nlargest(10)
```

```python
most_medals.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Total Medals')
plt.title('Countries with the Most Medals')
plt.show()
```

Task II: Which countries have won the most gold medals?


```python
most_gold_medals = df[df['Total Medals'] == 'Gold Medals'].groupby('Country')['Gold Medals'].count().nlargest(10)
```

```python
most_gold_medals.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Gold Medals')
plt.title('Countries with the Most Gold Medals')
plt.show()
```

Task III: Which countries have won the most medals in a particular sport?


```python
most_medals_sport = df.groupby(['Country', 'Sport'])['Total Medals'].count().reset_index()
most_medals_sport = most_medals_sport.groupby('Country').apply(lambda x: x[x['Total Medals'] == x['Total Medals'].max()])['Sport']
```

```python
most_medals_sport.value_counts().plot(kind='bar')
plt.xlabel('Sport')
plt.ylabel('Number of Countries')
plt.title('Countries with the Most Medals in a Sport')
plt.show()
```

Task IV: Which athletes have won the most medals?


```python
most_medals_athletes = df.groupby('Athlete')['Total Medals'].count().nlargest()
```

```python
most_medals_athletes.plot(kind='bar')
plt.xlabel('Athlete')
plt.ylabel('Number of Medals')
plt.title('Athletes with the Most Medals')
plt.show()
```

Task V: Which athletes have won the most gold medals?


```python
most_gold_medals_athlete = df[df['Gold Medals'] == 'Gold'].groupby('Athlete')['Gold Medals'].count().nlargest(10)
```

```python
most_gold_medals_athletes.plot(kind='bar')
plt.xlabel('Athlete')
plt.ylabel('Number of Gold Medals')
plt.title('Athlete with the Most Gold Medals')
plt.show()
```

Task VI: Which athletes have won the most medals in a particular sport?


```python
most_medals_sport_athletes = df.groupby(['Athlete', 'Sport'])['Total Medals'].count().reset_index()
most_medals_sport_athletes = most_medals_sport_athletes.groupby('Athlete').apply(lambda x: x[x['Total Medals'] == x['Total Medals'].max()])['Sport']
```

```python
most_medals_sport_athletes.value_counts().plot(kind='bar')
plt.xlabel('Sport')
plt.ylabel('Number of Athletes')
plt.title('Athletes with the Most Medals in a Sport')
plt.show()
```

Task VII: Which age group has won the most number of medals?


```python
most_medals_age = df.groupby('Age')['Total Medals'].count().nlargest(10)
```

```python
most_medals_age.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Number of Medals')
plt.title('Age Groups with the Most Medals')
plt.show()
```

Task VIII: Are there any trends in the gender of athletes who are winning medals?


```python
medal_gender_trends = df.groupby(['Year', 'Age'])['Total Medals'].count().unstack().fillna(0)
```

```python
medal_gender_trends.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Gender Trends of Medal Winners')
plt.legend(title='Age')
plt.show()
```
