## 1. Plot a comparison plot of following two expressions y=x^2+2x  , z=sin(y)



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-10, 10, 100)

# Calculate the values of y and z
y = x**2 + 2*x
z = np.sin(y)

# Plot the comparison
plt.plot(x, y, label='y = x^2 + 2x')
plt.plot(x, z, label='z = sin(y)')
plt.xlabel('x')
plt.ylabel('y, z')
plt.title('Comparison Plot: y = x^2 + 2x vs. z = sin(y)')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Assume that you are given with marks of 60 students of python course between 0-20 .Plot a histogram from this.

```python
import matplotlib.pyplot as plt

# Marks of the students
marks = [10]  # Replace [10] with the actual marks of the 60 students

# Plotting the histogram
plt.hist(marks, bins=10, range=(0, 20), edgecolor='black')

# Adding labels and title
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.title('Distribution of Marks in Python Course')

# Displaying the histogram
plt.show()
```

## 3. Using subplot, demonostrate curve for sin(x), cos (x) and tan(x)


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-np.pi, np.pi, 100)

# Calculate the values of sin(x), cos(x), and tan(x)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

# Create the subplot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))

# Plot sin(x)
ax1.plot(x, sin_x, label='sin(x)', color='blue')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_title('Sine Function')

# Plot cos(x)
ax2.plot(x, cos_x, label='cos(x)', color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.set_title('Cosine Function')

# Plot tan(x)
ax3.plot(x, tan_x, label='tan(x)', color='green')
ax3.set_xlabel('x')
ax3.set_ylabel('tan(x)')
ax3.set_title('Tangent Function')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

## 4. Using pie plot demonostrate the contribution of the runs scored by its batsman in an innings . (assume necessary data yourself).

```python
import matplotlib.pyplot as plt

# Runs scored by batsmen
batsmen_runs = {
    'Player 1': 56,
    'Player 2': 82,
    'Player 3': 32,
    'Player 4': 45,
    'Player 5': 23
    # Add more players and their runs as required
}

# Calculate the total runs
total_runs = sum(batsmen_runs.values())

# Calculate the contribution percentage for each batsman
contribution_percentages = [(runs / total_runs) * 100 for runs in batsmen_runs.values()]

# Get the batsmen names and their contribution percentages
batsmen_names = list(batsmen_runs.keys())

# Plotting the pie chart
plt.pie(contribution_percentages, labels=batsmen_names, autopct='%1.1f%%', startangle=90)

# Adding a title
plt.title('Contribution of Batsmen in an Innings')

# Displaying the pie chart
plt.show()
```
