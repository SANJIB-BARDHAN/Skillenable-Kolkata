## Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt

```

## Read the CSV file into a pandas DataFrame:


```python
df = pd.read_csv("E:\SKILLENABLE\INTERVIEW PREP\zomato_bang.csv")
df.head()

```

```python
df.columns
```

## Identify the most popular restaurants in the city.


```python
popular_restaurants = df.sort_values(by='votes', ascending=False).head(10)
popular_restaurants
```

```python
popular_restaurants[['name', 'votes']]
popular_restaurants
```

## Analyze the relationship between restaurant rating and price.


```python
plt.scatter(df['rate'], df['approx_cost(for two people)'])
plt.xlabel('Rating')
plt.ylabel('Price')
plt.title('Restaurant Rating vs Price')
plt.show()
```

## Identify the most popular cuisines in the city.


```python
popular_cuisines = df['cuisines'].value_counts().head(10)
popular_cuisines
```

```python
print(popular_cuisines)
```

```python
popular_cuisines.plot(kind='bar')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top 10 Popular Cuisines')
plt.show()
```

## Which restaurant has the highest number of votes?


```python
restaurant_highest_votes = df[df['votes'] == df['votes'].max()]['name'].iloc[0]
print(restaurant_highest_votes)

```

## How many restaurants belong to each restaurant type
(e.g., cafe, bar, bakery, etc.)?


```python
restaurant_type_counts = df['rest_type'].value_counts()
print(restaurant_type_counts)

```

## How many restaurant take online order.


```python
online_order_count = df[df['online_order'] == 'Yes'].shape[0]
print(online_order_count)

```

## Can we identify any popular menu items across different restaurants?


```python
popular_menu_items = df['menu_item'].value_counts().head(10)
print(popular_menu_items)

```
