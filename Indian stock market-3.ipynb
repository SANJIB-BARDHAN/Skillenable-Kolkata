```python
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline

```

# Load the dataset


```python
df = pd.read_excel(r"E:\SKILLENABLE\INTERVIEW PREP\Indian stock market final sheet Part 1.xlsx") 
df
```

# Calculate the growth rates by sub-sector


```python
growth_rates = df.groupby('Sub-Sector')['PE Ratio'].mean()
growth_rates
```

```python
# Plotting the growth rates
plt.figure(figsize=(12, 6))
growth_rates.plot(kind='bar')
plt.xlabel('Sub-Sector')
plt.ylabel('Average Growth Rate')
plt.title('Comparison of Growth Rates by Sub-Sector')
plt.xticks(rotation=60)
plt.show()
```

# Screen the companies by sub-sector


```python
screened_data = df.groupby('Sub-Sector')['Ticker'].count()
screened_data
```

```python
# Plotting the screened companies
plt.figure(figsize=(12, 6))
screened_data.plot(kind='bar')
plt.xlabel('Sub-Sector')
plt.ylabel('Number of Companies')
plt.title('Companies by Sub-Sector')
plt.xticks(rotation=60)
plt.show()
```

# Segregate the companies by Market Cap


```python
large_cap = df[df['Market Cap'] >= 10000]  # Replace 10000 with the threshold for large cap
mid_cap = df[(df['Market Cap'] >= 1000) & (df['Market Cap'] < 10000)]  # Replace 1000 and 10000 with the threshold for mid cap
small_cap = df[df['Market Cap'] < 1000]  # Replace 100 with the threshold for small cap

large_cap
mid_cap
small_cap
```

# Count the companies in each category


```python
cap_counts = [len(large_cap), len(mid_cap), len(small_cap)]
cap_categories = ['Large Cap', 'Mid Cap', 'Small Cap']

cap_counts
cap_categories
```

# Plotting the companies by Market Cap


```python
plt.figure(figsize=(8, 6))
plt.bar(cap_categories, cap_counts)
plt.xlabel('Market Cap Category')
plt.ylabel('Number of Companies')
plt.title('Companies by Market Cap Category')
plt.show()
```

# Select 10 random companies


```python
import random

random_companies = random.sample(df['Name'].unique().tolist(), 10)
random_companies
```

# Filter the data for the selected companies


```python
selected_data = df[df['Name'].isin(random_companies)]
selected_data
```

# Create a line chart for the selected companies


```python
plt.figure(figsize=(18, 6))
for company in random_companies:
    company_data = selected_data[selected_data['Name'] == company]
    plt.plot(company_data['Name'], company_data['Net Income'], label=company)

plt.xlabel('Name')
plt.ylabel('Net Income')
plt.title('Line Chart for 10 Random Companies')
plt.legend()
plt.show()
```
